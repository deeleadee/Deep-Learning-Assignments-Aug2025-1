**1. Write a Python program to create a bar graph that displays the population of five
different cities. Customize the bar graph by adding a title, labels for the axes, and
different colors for each bar.**
___

```python
import matplotlib.pyplot as plt
# Data
cities = ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix']
populations = [8419000, 3980000, 2716000, 2328000, 1690000]

#Code here:

colors = ['skyblue', 'lightgreen', 'salmon', 'plum', 'brown']

plt.figure(figsize=(10, 6))
plt.bar(cities, populations, color=colors)

plt.title('Population of Five Major U.S. Cities')
plt.xlabel('City')
plt.ylabel('Population')
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

**2. Write a Python program to create a pie chart that represents the market share
 of different smartphone brands. Label each wedge with the brand name and percentage, and
 highlight (explode) the wedge representing higher market share.**
 ___

```python
# Data
brands = ['Apple', 'Samsung', 'Huawei', 'Xiaomi', 'Oppo']
market_share = [27.5, 21.3, 15.8, 10.5, 8.3]

#Code here::

explode = [0.1 if share == max(market_share) else 0 for share in market_share]

plt.figure(figsize=(8, 8))
plt.pie(market_share, labels=brands, autopct='%1.1f%%', startangle=140, explode=explode, shadow=True)

plt.title('Smartphone Market Share by Brand')
plt.axis('equal')

plt.show()
```

**5. Write a Python program to create a line plot that shows the performance of three different students over five exams.Plot all three lines on the same graph, using different colors and markers for each line. Add a legend to identify each student.**
___

```python
# Data
exams = [1, 2, 3, 4, 5]
student1_scores = [80, 85, 88, 90, 92]
student2_scores = [78, 82, 85, 87, 89]
student3_scores = [75, 80, 82, 85, 88]

#Code here::
plt.figure(figsize=(10, 6))

plt.plot(exams, student1_scores, marker="p", color='blue', linestyle='-', label='Student 1')
plt.plot(exams, student2_scores, marker="P", color='green', linestyle='--', label='Student 2')
plt.plot(exams, student3_scores, marker="*", color='red', linestyle='-.', label='Student 3')

plt.title('Performance of Students Over 5 Exams')
plt.xlabel('Exam Number')
plt.ylabel('Score')
plt.xticks(exams)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```

**6. Write a Python program to create a stacked bar chart that represents the sales of three different products over four quarters. use different colors for each product, and add a legend.**
___

```python
# Data
quarters = ['Q1', 'Q2', 'Q3', 'Q4']
product1_sales = [150, 200, 250, 300]
product2_sales = [100, 150, 200, 250]
product3_sales = [50, 100, 150, 200]

#Code here::
plt.figure(figsize=(10, 6))

# Plot stacked bars
plt.bar(quarters, product1_sales, label='Product 1', color='skyblue')
plt.bar(quarters, product2_sales, bottom=product1_sales, label='Product 2', color='lightgreen')
# Calculate bottom for product3 by summing product1 and product2
bottom_product3 = [p1 + p2 for p1, p2 in zip(product1_sales, product2_sales)]
plt.bar(quarters, product3_sales, bottom=bottom_product3, label='Product 3', color='salmon')

# Customize
plt.title('Sales of Products Over 4 Quarters')
plt.xlabel('Quarter')
plt.ylabel('Sales')
plt.legend()
plt.grid(True, axis='y', linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```

**19. Using the Titanic dataset create a figure with two subplots: one subplot showing a histogram of the 'age' column with 50 bins and the second subplot showing a bar chart of the 'pclass' column. Ensure that both plots share the same x-axis.**
___

```python
#Code here:
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

axes[0].hist(df['age'].dropna(), bins=50, color='lightblue', edgecolor='black')
axes[0].set_title("Age Distribution")
axes[0].set_xlabel("Age")
axes[0].set_ylabel("Frequency")

pclass_counts = df['pclass'].value_counts().sort_index()
axes[1].bar(pclass_counts.index, pclass_counts.values, color='orange')
axes[1].set_title("Passenger Class Distribution")
axes[1].set_xlabel("Pclass")
axes[1].set_ylabel("Count")

plt.tight_layout()
plt.show()
```

```python
#Code here:
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

axes[0].plot(df['age'], color='teal')
axes[0].set_title("Age Line Plot")
axes[0].set_xlabel("Passenger Index")
axes[0].set_ylabel("Age")

sex_counts = df['sex'].value_counts()
axes[1].bar(sex_counts.index, sex_counts.values, color=['lightblue', 'pink'])
axes[1].set_title("Gender Distribution")
axes[1].set_xlabel("Sex")
axes[1].set_ylabel("Count")

plt.tight_layout()
plt.show()
```
